## Упражнение 1.

Построить графики плотности вероятности для 5 первых энергетических уровней гармонического осциллятора. Составить программы моделирования этих распределений методом исключения.

```python
def probability_density_quantum_oscillator(n: int, x: float):
    return (math.exp(- x ** 2) * hermite(n)(x) ** 2) / (math.sqrt(math.pi) * math.factorial(n) * 2 ** n)
```

```python
xs = np.linspace(-5, 5, 1000)

plt.scatter(x=xs, y=[probability_density_quantum_oscillator(0, x) for x in xs])
plt.show()
```
